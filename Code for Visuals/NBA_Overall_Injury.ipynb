```python
import pandas as pd
import numpy as np
import plotly.express as px


route='https://raw.githubusercontent.com/asingh2066/Visualization/main/injuries_2010-2020.csv'

data=pd.read_csv(route)

data.head()
```

```python
data.update(data['Team'].fillna(value='No team available', inplace=True))
data['year'] = pd.DatetimeIndex(data['Date']).year
data['month'] = pd.DatetimeIndex(data['Date']).month
data['year']=data.year.astype(str)
data['month']=data.month.astype(str)
data.head()
```

```python
fig = px.histogram(data, x="Team")
fig.update_layout(
   paper_bgcolor='rgb(255,255,255)',
   plot_bgcolor='rgb(255,255,255)',
    font_family="Helvetica",
    font_color="gray",
    title_font_family="Helvetica",
    title_font_color="gray",
    xaxis = { 
    'showgrid': False, 
    'zeroline': True, 
    'visible': True,
    
    },
    yaxis = { 
    'showgrid': False, 
    'zeroline': True, 
    'visible': True,
    
    }
    
)
fig.show()
```

```python
dataYear=data.groupby('year').size().reset_index(name='count')
dataYear=dataYear.sort_values('count', ascending=False)

fig = px.bar(dataYear,x='year',y='count')
fig.update_layout(
   paper_bgcolor='rgb(255,255,255)',
   plot_bgcolor='rgb(255,255,255)',
    font_family="Helvetica",
    font_color="gray",
    title_font_family="Helvetica",
    title_font_color="gray",
    xaxis = { 
    'showgrid': False, 
    'zeroline': True, 
    'visible': True,
    
    },
    yaxis = { 
    'showgrid': False, 
    'zeroline': True, 
    'visible': True,
    
    }
    
)
fig.show()
```
